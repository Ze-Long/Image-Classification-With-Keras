## Import Libraries and Packages

```python
import os
import numpy as np
import matplotlib.pyplot as plt

import keras
from keras.preprocessing.image import ImageDataGenerator
```

```python
os.listdir('concrete_data_week2')
```

## Construct an ImageDataGenerator Instance

Define a variable that represents the path to the folder containing the data

```python
dataset_dir = './concrete_data_week2'
```

#### Standard ImageDataGenerator

Define a standard ImageDataGenerator

```python
data_generator = ImageDataGenerator()
```

Use the <code>flow_from_directory</code> methods to loop through the images in batches.

```python
image_generator = data_generator.flow_from_directory(
    dataset_dir,
    batch_size=4,
    class_mode='categorical',
    seed=24
    )
```

Here, it found 40,000 images in total belonging to 2 classes.

#### Custom ImageDataGenerator

```python
data_generator = ImageDataGenerator(
    rescale=1./255
)
```

Proceed with defining the *image_generator* using the *flow_from_directory* method, just like before.

```python
image_generator = data_generator.flow_from_directory(
    dataset_dir,
    batch_size=4,
    class_mode='categorical',
    seed=24
    )
```

Explore the first batch using the *next* method, 

```python
first_batch_images = image_generator.next()[0]
second_batch_images = image_generator.next()[0]
third_batch_images = image_generator.next()[0]
fourth_batch_images = image_generator.next()[0]
fifth_batch_images = image_generator.next()[0]
```

Values are not integer values anymore, but scaled resolution since the original number are divided by 255.

## Visualize Batches of Images

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(20, 10)) 

ind = 0
for ax1 in axs:
    for ax2 in ax1: 
        image_data = first_batch_images[ind]
        ax2.imshow(image_data)
        ind += 1

fig.suptitle('First Batch of Concrete Images') 
plt.show()
```
